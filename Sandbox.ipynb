```python
    # import Part Detail
PartDetail = pd.read_csv('Part x WCentre.csv',index_col = 0)
PartDetail.head()
```

```python
    # import Order Detail
OrderDetail = pd.read_csv('FO Order list.csv')
OrderDetail.head()
```

```python
# clean data
OrderDetail.info()
```

```python
# manipulate data

# join tables
OrderDetail_wc = OrderDetail.merge(PartDetail, on='Part Number')
OrderDetail_wc.head()
```

```python
# Calculate FO Cycletime
OrderDetail_wc['Start'] = ConvertDate(OrderDetail_wc['FO Start'])
OrderDetail_wc['Finish'] = ConvertDate(OrderDetail_wc['FO Finish'])
OrderDetail_wc['CycleTime'] = OrderDetail_wc['Finish'] - OrderDetail_wc['Start']
OrderDetail_wc.head()
```
